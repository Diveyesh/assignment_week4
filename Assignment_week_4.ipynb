#Question-5

```python
#Inheritance is defined as the mechanism of inheriting the properties of the base class to the child class.
```

```python
#Single Inheritance
#Single inheritance enables a derived class to inherit properties from a single parent class,
#thus enabling code reusability and the addition of new features to existing code.
#Example is given below:-
```

```python
# Base class
class Parent:
  def func1(self):
    print("This function is in parent class.")

# Derived class
class Child(Parent):
  def func2(self):
    print("This function is in child class.")
```

```python
object = Child()
object.func1()
object.func2()
```

```python
#Multiple Inheritance:
#When a class can be derived from more than one base class this type of inheritance is called multiple inheritances.
#In multiple inheritances, all the features of the base classes are inherited into the derived class.
#Example is given below:-
```

```python
# Base class1
class Mother:
  def mother(self):
    print("Mother's name is Mrs. Geeta")

# Base class2
class Father:
  def father(self):
    print("Father's name is Mr. Suresh")

# Derived class
class Son(Mother, Father):
  def son(self):
    print("My name is Mr. Ramesh")
```

```python
s1 = Son()
s1.father()
s1.mother()
s1.son()
```

```python
#Multilevel Inheritance :
#In multilevel inheritance, features of the base class and the derived class are further inherited into the new derived class.
#Example is given below:-
```

```python
# Base class
class Grandfather:
  def grandfather_name(self):
    print("The Grandfather's name is Mr. Jignesh")

# Intermediate class
class Father(Grandfather):
  def father_name(self):
    print("The Father's name is Mr. Suresh")

# Derived class
class Son(Father):
  def son_name(self):
    print("My name is Mr. Sohan")
```

```python
s2 = Son()
s2.grandfather_name()
s2.father_name()
s2.son_name()
```

```python
#Hierarchical Inheritance:
#When more than one derived class are created from a single base this type of inheritance is called hierarchical inheritance.
#Example is given below:-
```

```python
# Base class
class Parent:
  def parent_name(self):
    print("Parent's name are Mr. and Mrs. Mohan and Asha")

# Derived class1
class Child1(Parent):
  def name_1(self):
    print("My name is Child 1")

# Derivied class2
class Child2(Parent):
  def name_2(self):
    print("My name is Child 2")
```

```python
object1 = Child1()
object2 = Child2()
object1.parent_name()
object1.name_1()
object2.parent_name()
object2.name_2()
```

```python
#Hybrid Inheritance:
#Inheritance consisting of multiple types of inheritance is called hybrid inheritance.
#Example is given below:-
```

```python
class School:
  def func1(self):
    print("This function is in school.")

class Student1(School):
  def func2(self):
    print("This function is in student 1. ")

class Student2(School):
  def func3(self):
    print("This function is in student 2.")

class Student3(Student1, School):
  def func4(self):
    print("This function is in student 3.")
```

```python
object = Student3()
object.func1()
object.func2()
```
